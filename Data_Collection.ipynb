```python
import pandas as pd
```

```python
# read the csv file
df = pd.read_csv('most_recent.csv')
df
```

```python
print(df.dtypes)
```

```python
# convert the domestic lifetime gross column to an integer
df['Domestic Lifetime Gross'] = df['Domestic Lifetime Gross'].astype(int)
df.head()
```

```python
# convert the production budget and profit columns to an integer
df['Production Budget'] = df['Production Budget'].astype(int)
df['Profit'] = df['Profit'].astype(int)
df.head()
```

```python
# create an updated csv file
df.to_csv('most_recent.csv', index=False)

```

```python
# replace all 0 values in the dataframe with blank cells
df = df.replace(0, '')
df.head()
```

```python
# create a new csv file
df.to_csv('most_recent_with_blanks.csv', index=False)
```
