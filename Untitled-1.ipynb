```python
import pandas as pd

```

```python
# make a dataframe from Movie Inflation Data Complete
df = pd.read_csv('Movie Inflation Data Complete.csv')
df.head()
```

```python
# make a dataframe from BoxOfficeMojo
df2 = pd.read_csv('BoxOfficeMojo.csv')
df2.head()
```

```python
# capitalize the first letter of each word in the title column
df2['Title'] = df2['Title'].str.title()
# remove all punctuation from the title column
df2['Title'] = df2['Title'].str.replace('[^\w\s]','')
df2.head()
```

```python
# capitalize the first letter of each word in the title column
df['Title'] = df['Title'].str.title()
# remove all punctuation from the title column
df['Title'] = df['Title'].str.replace('[^\w\s]','')
df.head()
```

```python
# merge the two dataframes on the title column with an outer join
df3 = pd.merge(df, df2, on='Title', how='outer')
df3.head()

```

```python
# create a csv file from the merged dataframe
df3.to_csv('merged.csv')
```

```python
# sort by rank_x
df3.sort_values(by=['Rank_x'], inplace=True)
df3.head()
```

```python
# change rank_x to Inflation_Rank
df3.rename(columns={'Rank_x': 'Inflation_Rank'}, inplace=True)
df3.head()
```

```python
# drop the year_y column
df3.drop('Year_y', axis=1, inplace=True)
df3.head()
```

```python
# change year_x to Year
df3.rename(columns={'Year_x': 'Year'}, inplace=True)
df3.head()
```

```python
# change rank_y to Worldwide_Rank
df3.rename(columns={'Rank_y': 'Worldwide_Rank'}, inplace=True)
df3.head()
```

```python
# convert the Inflation_Rank column to numeric and fill in the NaN values with 0
df3['Inflation_Rank'] = pd.to_numeric(df3['Inflation_Rank'], errors='coerce')
df3['Inflation_Rank'].fillna(0, inplace=True)
df3.head()

```

```python
# drop the decimals in the Inflation_Rank column
df3['Inflation_Rank'] = df3['Inflation_Rank'].astype(int)
df3.head()
```

```python
# convert the year column to numeric and fill in the NaN values with 0
df3['Year'] = pd.to_numeric(df3['Year'], errors='coerce')
df3['Year'].fillna(0, inplace=True)
df3.head()
```

```python
# drop the decimals in the Year column
df3['Year'] = df3['Year'].astype(int)
df3.head()
```

```python
# convert the worldwide_rank column to numeric and fill in the NaN values with 0
df3['Worldwide_Rank'] = pd.to_numeric(df3['Worldwide_Rank'], errors='coerce')
df3['Worldwide_Rank'].fillna(0, inplace=True)
df3.head()
```

```python
# drop the decimals in the Worldwide_Rank column
df3['Worldwide_Rank'] = df3['Worldwide_Rank'].astype(int)
df3.head()
```

```python
# put a $ in front of the Worldwide Lifetime Gross column
df3['Worldwide Lifetime Gross'] = '$' + df3['Worldwide Lifetime Gross'].astype(str)
df3.head()
```

```python
# add commas to the Worldwide Lifetime Gross column
df3['Worldwide Lifetime Gross'] = df3['Worldwide Lifetime Gross'].str.replace(r'(\d)(?=(\d\d\d)+(?!\d))', r'\1,')
df3.head()
```

```python
# drop the decimals in the Worldwide Lifetime Gross column
df3['Worldwide Lifetime Gross'] = df3['Worldwide Lifetime Gross'].str.replace(r'\.\d+', '')
df3.head()
```

```python
# put a $ in front of the domestic lifetime gross column
df3['Domestic Lifetime Gross'] = '$' + df3['Domestic Lifetime Gross'].astype(str)
df3.head()
```

```python
# add commas to the domestic lifetime gross column
df3['Domestic Lifetime Gross'] = df3['Domestic Lifetime Gross'].str.replace(r'(\d)(?=(\d\d\d)+(?!\d))', r'\1,')
df3.head()
```

```python
# drop the decimals in the domestic lifetime gross column
df3['Domestic Lifetime Gross'] = df3['Domestic Lifetime Gross'].str.replace(r'\.\d+', '')
df3.head()
```

```python
# put a $ in front of the foreign lifetime gross column
df3['Foreign Lifetime Gross'] = '$' + df3['Foreign Lifetime Gross'].astype(str)
df3.head()
```

```python
# add commas to the foreign lifetime gross column
df3['Foreign Lifetime Gross'] = df3['Foreign Lifetime Gross'].str.replace(r'(\d)(?=(\d\d\d)+(?!\d))', r'\1,')
df3.head()
```

```python
# drop the decimals in the foreign lifetime gross column
df3['Foreign Lifetime Gross'] = df3['Foreign Lifetime Gross'].str.replace(r'\.\d+', '')
df3.head()
```

```python
# make a new csv file from the cleaned dataframe
df3.to_csv('cleaned.csv')
```

```python
# remove domestic lifetime gross column
df3.drop('Domestic Lifetime Gross', axis=1, inplace=True)
df3.head()
```

```python
# change lifetime gross to domestic lifetime gross
df3.rename(columns={'Lifetime Gross': 'Domestic Lifetime Gross'}, inplace=True)
df3.head()
```

```python
# change the adj. lifetime gross to ajd. domestic lifetime gross
df3.rename(columns={'Adj. Lifetime Gross': 'Adj. Domestic Lifetime Gross'}, inplace=True)
df3.head()
```

```python
# create new csv file from the cleaned dataframe
df3.to_csv('cleaned2.csv')
```

```python
# drop duplicate titles
df3 = df3.drop_duplicates(subset='Title', keep='first')
df3.head()
```

```python
# create a new csv file from the cleaned dataframe
df3.to_csv('cleaned3.csv')
```

```python
# drop 0 in the Inflation_Rank column
df3 = df3[df3.Inflation_Rank != 0]
df3.head()
```

```python
# create a new csv file from the cleaned dataframe
df3.to_csv('cleaned4.csv')
df3.columns
```

```python
# reorganize the columns
df3 = df3[['Title', 'Year', 'Inflation_Rank', 'Adj. Domestic Lifetime Gross', 'Est. Num Tickets', 'Worldwide_Rank', 'Worldwide Lifetime Gross', 'Domestic Lifetime Gross', ' Domestic % ', 'Foreign Lifetime Gross', ' Foreign % ']]
df3.head()
```

```python
# rename domestic % and foreign % columns to remove the space
df3.rename(columns={' Domestic % ': 'Domestic %', ' Foreign % ': 'Foreign %'}, inplace=True)
df3.head()
```

```python
# multipy the domestic % and foreign % columns by 100
df3['Domestic %'] = df3['Domestic %'] * 100
df3['Foreign %'] = df3['Foreign %'] * 100
df3.head()
```

```python
# make all null values in the domestic % and foreign % columns blank
df3['Domestic %'].fillna('', inplace=True)
df3['Foreign %'].fillna('', inplace=True)
df3.head()
```

```python
# make all $nan values in the domestic lifetime gross and foreign lifetime gross and worldwide lifetime gross columns blank
df3['Domestic Lifetime Gross'].replace('$nan', '', inplace=True)
df3['Foreign Lifetime Gross'].replace('$nan', '', inplace=True)
df3['Worldwide Lifetime Gross'].replace('$nan', '', inplace=True)
df3.head()
```

```python
# make all 0 values in worldwide rank blank
df3['Worldwide_Rank'].replace(0, '', inplace=True)
df3.head()
```

```python
# change worldwide rank title
df3.rename(columns={'Worldwide_Rank': 'Worldwide Rank'}, inplace=True)
df3.head()
```

```python
# create a new csv file from the cleaned dataframe
df3.to_csv('cleaned5.csv')
```
