```python
import pandas as pd
```

```python
# Read the most_recent.csv file into a DataFrame
df = pd.read_csv('most_recent.csv')
```

```python
# create a new column called 'Profit' with unknown values for the movies
df['Profit'] = "unknown"
```

```python
# eliminate the decimals from the yer column
df['Year'] = df['Year'].astype(int)

```

```python
# make the domestic lifetime gross column an integer
df['Domestic Lifetime Gross'] = df['Domestic Lifetime Gross'].str.replace('$', '')
df['Domestic Lifetime Gross'] = df['Domestic Lifetime Gross'].str.replace(',', '')
df['Domestic Lifetime Gross'] = df['Domestic Lifetime Gross'].astype(int)
df.head()
```

```python
# make the production budget column an integer
df['Production Budget'] = df['Production Budget'].str.replace('$', '')
df['Production Budget'] = df['Production Budget'].str.replace(',', '')
df['Production Budget'] = df['Production Budget'].astype(int)
df.head()
```

```python
# Fill in the values for the profit column
df['Profit'] = df['Domestic Lifetime Gross'] - df['Production Budget']
df.head()
```

```python
# Add $ and commas to the profit, domestic lifetime gross, and production budget columns
df['Profit'] = df['Profit'].map('${:,.2f}'.format)
df['Domestic Lifetime Gross'] = df['Domestic Lifetime Gross'].map('${:,.2f}'.format)
df['Production Budget'] = df['Production Budget'].map('${:,.2f}'.format)
df.head()
```

```python
# Make all null values in the data frame equal to 0
df = df.fillna(0)
df
```

```python
# eliminate all commas and dollar signs from the data frame
df = df.replace({'\$': '', ',': ''}, regex=True)
df.head()
```

```python
# Eliminate the decimals from the worldwide rank
df['Worldwide Rank'] = df['Worldwide Rank'].astype(int)
df
```

```python
# update the csv file
df.to_csv('most_recent.csv', index=False)
```

```python
# Convert the adjusted domestic lifetime gross, estimated number of tickets, worldwide lifetime gross, domemstic lifetime gross, foreign lifetime gross, production budget, and profit columnsto a float and then an integer
df['Adj. Domestic Lifetime Gross'] = df['Adj. Domestic Lifetime Gross'].astype(float)
df['Adj. Domestic Lifetime Gross'] = df['Adj. Domestic Lifetime Gross'].astype(int) 
df['Est. Num Tickets'] = df['Est. Num Tickets'].astype(float)
df['Est. Num Tickets'] = df['Est. Num Tickets'].astype(int)
df['Worldwide Lifetime Gross'] = df['Worldwide Lifetime Gross'].astype(float)
df['Worldwide Lifetime Gross'] = df['Worldwide Lifetime Gross'].astype(int)
df['Domestic Lifetime Gross'] = df['Domestic Lifetime Gross'].astype(float)
df['Domestic Lifetime Gross'] = df['Domestic Lifetime Gross'].astype(int)
df['Foreign Lifetime Gross'] = df['Foreign Lifetime Gross'].astype(float)
df['Foreign Lifetime Gross'] = df['Foreign Lifetime Gross'].astype(int)
df['Production Budget'] = df['Production Budget'].astype(float)
df['Production Budget'] = df['Production Budget'].astype(int)
df['Profit'] = df['Profit'].astype(float)
df['Profit'] = df['Profit'].astype(int)
df.head()

```
