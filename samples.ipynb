```python
# Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
from scipy.stats import linregress
import numpy as np
```

```python
# make a dataframe from csv
df = pd.read_csv('most_recent.csv')
df.head(10)
```

```python
#create top 10 adjusted for inflation with movies as y axis

movies = ['Gone With The Wind', 'Star Wars Episode IV: A New Hope', 'The Sound Of Music', 'ET: The Extra Terrestrial', 'Titanic', 'The Ten Commandments', 'Jaws', 'Doctor Zhivago', 'The Exorcist', 'Snow White And The Seven Dwarfs']
revenue = [1895421694, 1668989715, 1335086324, 1329174791, 1270101626, 1227470000, 1200856389, 1163149635, 1036314504, 1021330000]
years = [1939, 1977, 1965, 1982, 1997, 1956, 1975, 1965, 1973, 1937]

#Create the bar chart
fig, ax = plt.subplots(figsize=(20,30.5))
ax.bar(years, revenue)

#set the x-axis ticks and labels
ax.set_xticks(years)
ax.set_xticklabels(years)

#Set the y-axis label
ax.set_ylabel('Revenue')

#Set the tick labels
ax.set_yticks(revenue)
ax.set_yticklabels(movies)

#show the plot
ax.set_title('Top 10 Movies Adjusted Domestic Gross Lifetime for Inflation')
# Save the figure
plt.savefig("output_data/Fig1.png")
plt.show()

```

```python
#create linear regression for budget vs. domestic lifetime gross
x_values = df['Production Budget']/1000000
y_values = df['Domestic Lifetime Gross']/1000000
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(5.8,0.8),fontsize=15,color="red")
plt.xlabel('Production Budget ($ in millions)')
plt.ylabel('Domestic Lifetime Gross ($ in millions)')
plt.title('Production vs. Budget')
print(f"The r-squared is: {rvalue**2}")
# Save the figure
plt.savefig("output_data/Fig5.png")

plt.show()
```

```python
#create linear regression for budget vs. profit
x_values = df['Profit']/100000
y_values = df['Production Budget']/100000
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6.0,0.8),fontsize=15,color="red")
plt.xlabel('Production Budget ($ in millions)')
plt.ylabel('Profit ($ in millions)')
plt.title('Profit vs. Production Budget')
print(f"The r-squared is: {rvalue**2}")
# Save the figure
plt.savefig("output_data/Fig6.png")

plt.show()
```
